#1st we take a random XOR dataset than we are performing ANN


```python
import numpy as np
```

#xor dataset

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

#initialize weight

####Input Layer  : 2 neurons
####Hidden Layer : 3 neurons
####Output Layer : 1 neuron
####Weight matrices :W1 (Input → Hidden)shape = (2 × 3)
####W2 (Hidden → Output) shape = (3 × 1)

```python
np.random.seed(1)
w1 = np.random.randn(2,3)
#Input Layer : 2 neurons
#Hidden Layer : 3 neurons
w2 = np.random.randn(3,1)
#W2 (Hidden → Output) shape = (3 × 1)
```

#Activation

```python
def sigmod(x):
  return 1/(1+np.exp(-x))  #forword pro

def sigmod_derivative(x):   #backword pro(Gradient[loss propagation])
  return x*(1-x)
```

##Used in backpropagation to calculate gradient. It tells how much weights should change.
# training

```python
for i in range(10000):
   # -------- Feedforward --------
  Hidden_layer = sigmod(np.dot(x,w1))#→ Hidden Layer atrix multiplication:
                                    #(4×2) dot (2×3) = (4×3)Then apply sigmoid activation.

  Output = sigmod(np.dot(Hidden_layer,w2))#Hidden → Output Matrix multiplication:
  error = y-Output
  print(error)
  d_output = error*sigmod_derivative(Output)
  error_hidden = d_output.dot(w2.T)
  d_hidden = error_hidden*sigmod_derivative(Hidden_layer)
  w2 += Hidden_layer.T.dot(d_output)
  w1 += x.T.dot(d_hidden)
```
