```python
# Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical

# Training text (small example dataset)
text = "hello world"
```

```python
# Find all unique characters in the text
chars = sorted(list(set(text)))

# Create dictionary to convert characters to numbers
char_to_int = {c:i for i,c in enumerate(chars)}

# Create reverse dictionary (number → character)
int_to_char = {i:c for i,c in enumerate(chars)}
```

```python
# Length of input sequence
seq_length = 3

# Input and output lists
X = []
y = []
```

```python
# Create training sequences
for i in range(len(text) - seq_length):

    # Input characters
    seq_in = text[i:i+seq_length]

    # Next character to predict
    seq_out = text[i+seq_length]

    # Convert characters to numbers
    X.append([char_to_int[c] for c in seq_in])
    y.append(char_to_int[seq_out])
```

```python
# Convert lists to numpy arrays
X = np.array(X)

# Reshape for RNN input (samples, time steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))

# Normalize input values
X = X / float(len(chars))

# Convert output to categorical format
y = to_categorical(y)

# Build RNN model
model = Sequential()

# Add SimpleRNN layer with 32 neurons
model.add(SimpleRNN(32, input_shape=(seq_length,1)))

# Output layer with softmax activation
model.add(Dense(y.shape[1], activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam')

# Train the model
model.fit(X, y, epochs=500, verbose=0)

# Starting text for generation
seed = "hel"

# Convert seed to numbers
pattern = [char_to_int[c] for c in seed]

print("Generated text:")

# Generate 10 characters
for i in range(10):

    # Prepare input
    x = np.array(pattern)
    x = x.reshape((1, len(pattern), 1))
    x = x / float(len(chars))

    # Predict next character
    prediction = model.predict(x, verbose=0)

    # Choose character with highest probability
    index = np.argmax(prediction)

    # Convert number to character
    result = int_to_char[index]

    # Print generated character
    print(result, end="")

    # Update sequence
    pattern.append(index)
    pattern = pattern[1:]
    #Your output:Generated text:lo worldll
#The model is generating characters based on learned patterns.

#Training text:

#hello world

#Patterns learned by the RNN:  hel → l ell → o llo → space lo  → w
#So when the seed is: hel  The model predicts: lo world

#After that it continues generating characters such as: l l

#because it keeps predicting based on probabilities.
```
