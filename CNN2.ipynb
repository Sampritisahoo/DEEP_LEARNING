```python
# Import TensorFlow library for deep learning
import tensorflow as tf

# Import Keras modules for building CNN layers
from tensorflow.keras import layers, models

# Import dataset utilities
import tensorflow_datasets as tfds

# ============================================================
# STEP 1: LOAD DATASET
# ============================================================

# Load Cats vs Dogs dataset from TensorFlow datasets
# Dataset automatically downloads images
(dataset, info) = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:]'],  # 80% training, 20% testing
    as_supervised=True,
    with_info=True
)

train_dataset, test_dataset = dataset

print("Dataset Loaded Successfully")
```

```python
# ============================================================
# STEP 2: IMAGE PREPROCESSING
# ============================================================

# Define image size
#IMG_SIZE = 150
IMG_SIZE = 80

# Function to resize and normalize images
def preprocess(image, label):

    # Resize image to 150x150 pixels
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))

    # Normalize pixel values (0-255 → 0-1)
    image = image / 255.0

    return image, label

# Apply preprocessing
train_dataset = train_dataset.map(preprocess)
test_dataset = test_dataset.map(preprocess)

# Batch dataset for training
train_dataset = train_dataset.batch(32)
test_dataset = test_dataset.batch(32)
```

```python
# ============================================================
# STEP 3: BUILD CNN MODEL
# ============================================================

import tensorflow as tf
from tensorflow.keras import layers, models

# Build CNN model
model = models.Sequential()

# First convolution
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(80,80,3)))
model.add(layers.MaxPooling2D((2,2)))

# Second convolution
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))

# Third convolution
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))

# Instead of Flatten (prevents shape mismatch)
model.add(layers.GlobalAveragePooling2D())

# Fully connected layer
model.add(layers.Dense(128,activation='relu'))

# Output layer (Cat vs Dog)
model.add(layers.Dense(1,activation='sigmoid'))

model.summary()
```

```python
# ============================================================
# STEP 4: COMPILE MODEL
# ============================================================

# Configure training process
model.compile(
    optimizer='adam',                # Optimization algorithm
    loss='binary_crossentropy',      # Loss function for two classes
    metrics=['accuracy']             # Evaluation metric
)
```

```python
# ============================================================
# STEP 5: TRAIN MODEL
# ============================================================

# Train CNN using training dataset
history = model.fit(
    train_dataset,
    epochs=2,
    validation_data=test_dataset
)
```

```python
# ============================================================
# STEP 6: EVALUATE MODEL
# ============================================================

# Test CNN performance
test_loss, test_acc = model.evaluate(test_dataset)

print("Test Accuracy:", test_acc)
```

```python
# ============================================================
# STEP 7: MAKE PREDICTIONS
# ============================================================

# Take one batch of images
for images, labels in test_dataset.take(1):

    # Predict class probabilities
    predictions = model.predict(images)

    print("Prediction probability:", predictions[0])
    print("Actual label:", labels[0])

# ============================================================
# OUTPUT INTERPRETATION
# ============================================================

# Output close to 1 → Dog
# Output close to 0 → Cat

# Example:
# Prediction = 0.92 → Dog
# Prediction = 0.10 → Cat

# ============================================================
# END OF CNN CAT vs DOG MODEL
# ============================================================
```
