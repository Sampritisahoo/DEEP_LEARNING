```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

#library

#load dataset

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"

columns = [
'Pregnancies','Glucose','BloodPressure','SkinThickness',
'Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'
]

data = pd.read_csv(url, names=columns)

data.head()
```

#Separate feature and target

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

#tarin test split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

#noemalize Data

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

#Build ANN model

```python
model = keras.Sequential([
    keras.layers.Dense(16, activation='relu', input_shape=(8,)),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

#compiler model

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
) #Optimizer controls how model weights are updated during training. Adam = Adaptive Moment Estimation :It combines:Momentum,

#RMSProp Advantages: fast convergence. good performance for deep learning ,Concept: Weight_new = Weight_old - learning_rate Г— gradient

#Optimizer decides how much the weights should change.
# Loss function measures prediction error. Binary cross-entropy is used when the problem is binary classification. In this dataset:

#Class	Meaning:0	No Diabetes,1	Diabetes

#Formula:Loss = -[ y log(p) + (1-y) log(1-p) ]

#Where: y = true label,p = predicted probability,Lower loss в†’ better model.
```

#Train Model

```python
history = model.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.2
) #Epoch = one full pass through the dataset. Example: Epoch	Meaning
#1	dataset seen once
#50	dataset seen 50 times

#More epochs в†’ better learning (but too many may cause overfitting).
```

#Evaluted model

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
accuracy = model.evaluate(x_test, y_test)
print("accuracy:", accuracy)
```
