##Problem

##Train an RNN to learn a sequence pattern.
##ex sequence:
##1-->2
##2-->3
##3-->4
##4-->5
##4 input are present which is sequentialy attach each other

```python

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Input
```

```python
# Dataset
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 3, 4, 5])
```

```python
# Normalize
X = X / 10
y = y / 10
```

```python
# Reshape for RNN(RNN requires 3-dimensional input data.)
X = X.reshape((4,1,1))
```

```python
# Model(Creates a Sequential neural network where layers are stacked in order.)
model = Sequential()
model.add(Input(shape=(1,1)))
model.add(SimpleRNN(10, activation='tanh'))
model.add(Dense(1))
```

```python
#Parameters: 10,Number of hidden neurons.activation='tanh'Activation function used in the RNN. input_shape=(1,1) Input sequence format
# Compile
model.compile(optimizer='adam', loss='mse')
```

```python
# Train
model.fit(X, y, epochs=1000, verbose=0)
```

```python

# Test(samples=1, time steps=1, features=1)
test = np.array([[5]]) / 10
test = test.reshape((1,1,1))
prediction = model.predict(test)
# normalized output

print("Predicted value:", prediction*10)
```
